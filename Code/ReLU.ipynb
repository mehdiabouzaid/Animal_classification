# Imports

```python
from numpy import linspace
from numpy import maximum
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
```

# ReLU

```python
axis = linspace(-10,10,500)
f = maximum(0,axis)
```

# Bias

```python
bias = 5
```

# Weighted sum

$$ Let \quad weighted\_sum = \sum_{i=1}^n w_i \cdot x_i = w^T \cdot x = -2.5 $$

```python
weighted_sum = -2.5
```

# Plot

```python
figure = plt.figure(figsize=(15,15))

plot1 = plt.subplot(2,2,1)
plt.plot(axis,f)
plt.grid(color='gray', linestyle='-', linewidth=0.5)
plt.title('Without bias')
plt.xlabel('axis')
plt.ylabel('f(axis)')
plt.scatter(weighted_sum,0,marker="o",color="r",s=400)
plt.scatter(weighted_sum,0,marker="s",color="y",s=100)

# Légende
symbol1 = mlines.Line2D([], [], color='b', marker='', linestyle='solid',
                          markersize=10, label='ReLU')
symbol2 = mlines.Line2D([], [], color='r', marker='o', linestyle='None',
                          markersize=10, label='weighted_sum')
symbol3 = mlines.Line2D([], [], color='y', marker='s', linestyle='None',
                          markersize=10, label='ReLU(weighted_sum)')
plot1.legend(handles=[symbol1, symbol2, symbol3])

plot2 = plt.subplot(2,2,2)
plt.plot(axis,f)
plt.grid(color='gray', linestyle='-', linewidth=0.5)
plt.title('With bias')
plt.xlabel('axis')
plt.ylabel('f(axis)')
plt.scatter(weighted_sum+bias,0,marker="o",color="r",s=400)
plt.scatter(weighted_sum+bias,2.5,marker="s",color="y",s=100)

symbol2 = mlines.Line2D([], [], color='r', marker='o', linestyle='None',
                          markersize=10, label='weighted_sum+bias')
symbol3 = mlines.Line2D([], [], color='y', marker='s', linestyle='None',
                          markersize=10, label='ReLU(weighted_sum+bias)')
plot2.legend(handles=[symbol1, symbol2, symbol3])

plt.show()
```
